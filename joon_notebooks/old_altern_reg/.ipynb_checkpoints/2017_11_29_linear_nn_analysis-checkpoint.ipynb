```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import audtorch.metrics.functional as audtorch
import time
import numpy as np
from tqdm import tnrange, tqdm_notebook
from sklearn.decomposition import PCA
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages


```

```python
"""
Linear Neural Net Parameters:
"""

k_dim = 2
p_dim = 95*146
f_dim = 2
t_dim = 50
batch_size = 32
unit_no = 2094

save_dir = "/ssd/joon/2017_11_29_ns/yass/reduced_rank/"
```

```python


class LinearNet(nn.Module):
    def __init__(self, unit_no, t_dim, k_dim, f_dim, p_dim):
        super(LinearNet, self).__init__()
        self.unit_no = unit_no
        self.t_dim = t_dim
        self.k_dim = k_dim
        self.f_dim = f_dim
        self.p_dim = p_dim
        
        self.V = nn.Linear(t_dim, k_dim).cuda()
        self.U_list = nn.ModuleList([nn.Linear(k_dim, f_dim).cuda() for i in range(unit_no)]).cuda()
        self.W_list = nn.ModuleList([nn.Linear(f_dim, p_dim).cuda() for i in range(unit_no)]).cuda()
        
        """
        self.V.weight = torch.nn.Parameter(V_init)
        
        for i in range(unit_no):
            self.U_list[i].weight = torch.nn.Parameter(U_init[2*i : 2*(i+1),  : ])
            self.W_list[i].weight = torch.nn.Parameter(W_init[ 2*i : 2*(i+1) , :].T)
        """
        
    def forward(self, S):
        self.batch = S.shape[0]
        
        ###### (50n) --> (kn) ####
        VS = torch.empty(self.batch, self.k_dim * self.unit_no).cuda()
        
        for i in range(self.unit_no):
            S_i = S[:, self.t_dim*i : self.t_dim*(i+1)].float().cuda()
            VS_i = self.V(S_i)
            VS[:, self.k_dim*i : self.k_dim*(i+1)] = VS_i
            
        ###### (kn) --> (2n) #####
        UVS = torch.empty(self.batch, self.f_dim * self.unit_no).cuda()
            
        for i in range(self.unit_no):
            VS_i = VS[:, self.k_dim*i : self.k_dim*(i+1)].float().cuda()
            U_i_func = self.U_list[i]
            UVS_i = U_i_func(VS_i)
            UVS[:, self.f_dim*i : self.f_dim*(i+1)] = UVS_i
        
        ##### (2n) --> (p) #####
        I = torch.zeros(self.batch, self.p_dim).cuda()
 
        for i in range(self.unit_no):
            UVS_i = UVS[:, self.f_dim*i : self.f_dim*(i+1)].float().cuda()
            W_i_func = self.W_list[i]
            I_i = W_i_func(UVS_i)
            I += I_i
        
        return I.float()
```

```python
model = LinearNet(unit_no, t_dim, k_dim, f_dim, p_dim)
model.load_state_dict(torch.load("/ssd/joon/2017_11_29_ns/yass/reduced_rank/yass_I2_reduced_lnn_model.pt"))
model.eval()
```

# Extracting Temporal and Spatial Weights

```python
V = model.V.weight
#UV_on_trained = torch.empty(unit_no, t_dim)
#UV_off_trained = torch.empty(unit_no, t_dim)
UV_trained = torch.empty(unit_no*2, t_dim)
U_trained = torch.empty((unit_no*2, k_dim))

for i in tnrange(unit_no):
    U_n = model.U_list[i].weight
    Un_Vn = torch.matmul(U_n, V)
    
    UV_trained[2*i : 2*(i+1), :] = Un_Vn
    U_trained[2*i : 2*(i+1), :] = U_n
    
UV_trained = UV_trained.detach().cpu().numpy()
U_trained = U_trained.detach().cpu().numpy()

W = torch.empty(unit_no*2, p_dim)

for i in tnrange(unit_no):
    W_n = model.W_list[i].weight
    W[i*2 : (i+1)*2 , :] = W_n.T

W = W.detach().cpu().numpy()

V = V.detach().cpu().numpy()

save_dir = "/ssd/joon/2017_11_29_ns/yass/weights/"

np.save(os.path.join(save_dir, "yass_I2_lnn_UV_weights.npy"), UV_trained)
np.save(os.path.join(save_dir, "yass_I2_lnn_W_weights.npy"), W)
np.save(os.path.join(save_dir, "yass_I2_lnn_V_weights.npy"), V)
np.save(os.path.join(save_dir, "yass_I2_lnn_U_weights.npy"), U_trained)
```

```python
def plot_uv_weights(UV_file, plot_unit_no, index, sorted_units_file):
    UV_weights = np.load(UV_file)
    sorted_units = np.load(sorted_units_file)
    
    unit_ids = sorted_units[index*plot_unit_no : plot_unit_no*(index+1)]
    weight_ids = np.empty((2*plot_unit_no))
    weight_ids = weight_ids.astype(int)
    
    for i in range(plot_unit_no):
        weight_ids[i*2] = unit_ids[i]*2
        weight_ids[i*2+1] = unit_ids[i]*2+1
    
    plot_UV = UV_weights[weight_ids , :]
    UV_plot = plt.imshow(plot_UV)
    plt.xlabel("50 Time bins (10ms each)")
    plt.ylabel("2 Temporal Filters (ON,OFF) for 50 units")
    plt.yticks(np.arange(0,101,4))
    plt.xticks(np.arange(0,51,10))
    plt.colorbar(UV_plot)
    
    
def multi_plot_uv_weights(UV_file, plot_unit_no, index_no, sorted_units_file, save_dir):
    
    with PdfPages(os.path.join(save_dir, "yass_I2_lnn_UV_weights.pdf")) as pdf:
        for i in tnrange(index_no):
            fig = plot_uv_weights(UV_file, plot_unit_no, i, sorted_units_file)
            pdf.savefig(fig)
            plt.close()
```

```python
UV_file = "/ssd/joon/2017_11_29_ns/yass/weights/yass_I2_lnn_UV_weights.npy"
plot_unit_no = 50
save_dir = "/ssd/joon/2017_11_29_ns/figures/"
index_no = 20
sorted_units_file = "/ssd/joon/2017_11_29_ns/yass/yass_weight_sorted_units.npy"

multi_plot_uv_weights(UV_file, plot_unit_no, index_no, sorted_units_file, save_dir)
```

```python
def plot_w_weights(lnn_W_file, lin_reg_W_file, save_dir):
    lnn_W = np.load(lnn_W_file)
    linreg_W = np.load(lin_reg_W_file)
    unit_no = lnn_W.shape[0]//2
    p_dim = lnn_W.shape[1]
    
    abs_max_array = np.empty((unit_no))
    
    for i in tnrange(unit_no):
        abs_max = np.max(np.abs(lnn_W[i*2 : (i+1)*2 , :]))
        abs_max_array[i] = abs_max
        
    abs_max_ids = np.argsort(abs_max_array)
    abs_max_ids = np.flip(abs_max_ids)
    
    for i in tnrange(10):
        unit = abs_max_ids[i]
        fig, axs = plt.subplots(nrows = 2, ncols =2, figsize = (8,6), dpi = 80)
        on_lnn = axs[0,0].imshow(lnn_W[unit*2,:].reshape((95,146)))
        off_lnn = axs[1,0].imshow(lnn_W[unit*2 + 1,:].reshape((95,146)))
        
        on_linreg = axs[0,1].imshow(linreg_W[unit*2,:].reshape((95,146)))
        off_linreg = axs[1,1].imshow(linreg_W[unit*2 + 1,:].reshape((95,146)))
        
        axs[0,0].axis("off")
        axs[0,0].set_title("LNN (I2) Index" + str(i) + " Unit" + str(unit) + " ON")
        axs[1,0].axis("off")
        axs[1,0].set_title("LNN (I2) Index" + str(i) + " Unit" + str(unit) + " OFF")
        plt.colorbar(on_lnn, ax=axs[0,0], shrink = 0.7)
        plt.colorbar(off_lnn, ax=axs[1,0], shrink = 0.7)
        
        axs[0,1].axis("off")
        axs[0,1].set_title("LinReg Index" + str(i) + " Unit" + str(unit) + " ON")
        axs[1,1].axis("off")
        axs[1,1].set_title("LinReg Index" + str(i) + " Unit" + str(unit) + " OFF")
        plt.colorbar(on_linreg, ax=axs[0,1], shrink = 0.7)
        plt.colorbar(off_linreg, ax=axs[1,1], shrink = 0.7)
   
        fig.tight_layout()
        fig.savefig(os.path.join(save_dir, "index" + str(i) + "_unit" +str(unit)) + "_lnn_W_weights.png")
        plt.close()
```

```python
lnn_W_file = "/ssd/joon/2017_11_29_ns/yass/weights/yass_I2_lnn_W_weights.npy"
lin_reg_W_file = "/ssd/joon/2017_11_29_ns/yass/weights/yass_smooth_4_weights.npy"
save_dir = "/ssd/joon/2017_11_29_ns/yass/lnn_W_weights/"

plot_w_weights(lnn_W_file, lin_reg_W_file, save_dir)
```

```python
def comb_img_corr(one_file, two_file, three_file, four_file, test_file, save_dir):
    one = np.load(one_file)
    two = np.load(two_file)
    three = np.load(three_file)
    four = np.load(four_file)
    test = np.load(test_file)

    image_no = test.shape[0]
    
    comb_corr_array = np.empty((image_no, 4))  
    
    for i in tnrange(image_no):
        one_img = one[i,:]
        two_img = two[i,:]
        three_img = three[i,:]
        four_img = four[i,:]
        test_img = test[i,:]
        
        one_corr = np.corrcoef(one_img.reshape(-1), test_img.reshape(-1))[0][1]
        two_corr = np.corrcoef(two_img.reshape(-1), test_img.reshape(-1))[0][1]
        three_corr = np.corrcoef(three_img.reshape(-1), test_img.reshape(-1))[0][1]
        four_corr = np.corrcoef(four_img.reshape(-1), test_img.reshape(-1))[0][1]
        
        comb_corr_array[i,0] = one_corr
        comb_corr_array[i,1] = two_corr
        comb_corr_array[i,2] = three_corr
        comb_corr_array[i,3] = four_corr
        
    np.save(os.path.join(save_dir, "yass_I2_k7_k10_lnn_corr.npy"), comb_corr_array)
```

```python
one_file = "/ssd/joon/2017_11_29_ns/yass/decoded/yass_smooth_test_decoded.npy"
two_file = "/ssd/joon/2017_11_29_ns/yass/decoded/yass_I2_lnn_test_decoded.npy"
three_file = "/ssd/joon/2017_11_29_ns/yass/decoded/yass_k7_lnn_test_decoded.npy"
four_file = "/ssd/joon/2017_11_29_ns/yass/decoded/yass_k10_lnn_test_decoded.npy"
test_file = "/ssd/joon/2017_11_29_ns/images/smooth_test_images.npy"
save_dir = "/ssd/joon/2017_11_29_ns/yass/test_corr/"

comb_img_corr(one_file, two_file, three_file, four_file, test_file, save_dir)
```
